```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_PATH = 'data/customer_data.csv'
```

```python
# Load data
df = pd.read_csv(DATA_PATH)
df
```

```python
# Give min and max age
min_age = df['age'].min()
max_age = df['age'].max()
print(f'Min age: {min_age}')
print(f'Max age: {max_age}')
```

```python
# Same for income
min_income = df['income'].min()
max_income = df['income'].max()
print(f'Min income: {min_income}')
print(f'Max income: {max_income}')
```

```python
# Remove rows where income is above 1 000 000
df = df[df['income'] < 1_000_000]
df
```

```python
# Save
df.to_csv('data/customer_data_clean.csv', index=False)
```
