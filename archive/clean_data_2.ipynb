```python
import numpy as np
import pandas as pd
```

```python
DATA_PATH = "data/customer_data.csv"
```

```python
# Load the data
data = pd.read_csv(DATA_PATH)
```

```python
data
```

```python
data.dtypes[:10]
```

```python
# Remove customer_id and new_customer columns
data = data.drop(columns=["customer_id", "new_customer"])
```

```python
data
```

```python
np.unique(data["province_name"])
```

```python
# Remove commas from the province_name column
data["province_name"] = data["province_name"].str.replace(",", "")
```

```python
np.unique(data["province_name"])
```

```python
# Convert has_payroll to int
data["has_payroll"] = data["has_payroll"].astype(int)
data
```

```python
# Rename has_pension.1 to has_pension_personal
data = data.rename(columns={"has_pension.1": "has_pension_personal"})
# Convert has_pension_personal to int
data["has_pension_personal"] = data["has_pension_personal"].astype(int)
```

```python
data
```

```python
# Save data
data.to_csv("data/customer_data_clean.csv", index=False)
```
